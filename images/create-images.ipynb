```python
p = ggplot(data, aes('Bwt', 'Hwt')) + \
    geom_point(colour='steelblue', size=30) + \
    scale_x_continuous(limits = [1.9, 4])
    
p = p + xlab("Weight (kg)") + ylab("Height (cm)")# + geom_text(size=14)
p = p + geom_smooth(method="lm",se=False, size=3, colour="violet") + theme_bw()

ggsave(filename="cat_regression.jpg", width=15, height=7, units='in', plot=p)
```
